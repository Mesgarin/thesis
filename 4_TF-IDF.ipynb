```python
import os
import docx
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Function to read content from a docx file
def read_docx(file_path):
    doc = docx.Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

# Function to preprocess text (remove punctuation and convert to lowercase)
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()
    # Remove numbers and punctuation
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text

# Directory path containing the .docx files
directory = 'C:/Users/Nastaran/Desktop/UniversityConstructor/semester4/Final 08042025/Data/Similarity'

# List to store the text content of the files
texts = []
filenames = []

# Read all docx files from the directory
for filename in os.listdir(directory):
    if filename.endswith(".docx"):
        file_path = os.path.join(directory, filename)
        text = read_docx(file_path)
        texts.append(preprocess_text(text))  # Preprocess the text
        filenames.append(filename)

# Compute TF-IDF vectors for the texts
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(texts)

# Compute cosine similarity between all document pairs
similarity_matrix = cosine_similarity(tfidf_matrix)

# Create a DataFrame to store the results
similarity_results = []

# Display cosine similarity between the files and store in a DataFrame
for i, filename in enumerate(filenames):
    for j in range(i + 1, len(filenames)):
        similarity = similarity_matrix[i, j]
        similarity_results.append({
            'File 1': filenames[i],
            'File 2': filenames[j],
            'Similarity': round(similarity, 4)
        })

# Convert the results to a DataFrame
similarity_df = pd.DataFrame(similarity_results)

# Display the similarity DataFrame
print(similarity_df)

# Save the results to an Excel file
output_path = "C:/Users/Nastaran/Desktop/UniversityConstructor/semester4/Final 08042025/Data/docx_similarity_results.xlsx"
similarity_df.to_excel(output_path, index=False)
```

```python
import os
import docx
import re
from sentence_transformers import SentenceTransformer, util
import pandas as pd

# Function to read content from a docx file
def read_docx(file_path):
    doc = docx.Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

# Function to preprocess text (remove punctuation and convert to lowercase)
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text

# Load Sentence-BERT model
model = SentenceTransformer('paraphrase-distilroberta-base-v1')

# Directory containing DOCX files
directory = 'C:/Users/Nastaran/Desktop/UniversityConstructor/semester4/Final 08042025/Data/Similarity'

# Read files
texts = []
filenames = []

for filename in os.listdir(directory):
    if filename.endswith(".docx"):
        file_path = os.path.join(directory, filename)
        text = read_docx(file_path)
        texts.append(preprocess_text(text))
        filenames.append(filename)

# Create similarity results
similarity_results = []

# Compute cosine similarity between the files and sentences
for i, filename1 in enumerate(filenames):
    for j in range(i + 1, len(filenames)):
        text1 = texts[i]
        text2 = texts[j]

        # Split the documents into sentences
        sentences1 = text1.split('\n')
        sentences2 = text2.split('\n')

        # Encode sentences to vectors using Sentence-BERT
        sentence_embeddings1 = model.encode(sentences1, convert_to_tensor=True)
        sentence_embeddings2 = model.encode(sentences2, convert_to_tensor=True)

        # Calculate pairwise cosine similarity between sentences
        cosine_similarities = util.pytorch_cos_sim(sentence_embeddings1, sentence_embeddings2)

        # Iterate over the sentence pairs and find the highest similarity
        for k, sentence1 in enumerate(sentences1):
            for l, sentence2 in enumerate(sentences2):
                similarity = cosine_similarities[k][l].item()

                # If similarity is above a threshold, store the sentence pair and similarity
                if similarity > 0.5:  # You can adjust the threshold
                    similarity_results.append({
                        'File 1': filenames[i],  # Use filenames[i] instead of filename1
                        'File 2': filenames[j],  # Use filenames[j] instead of filename2
                        'Sentence 1': sentence1,
                        'Sentence 2': sentence2,
                        'Similarity': round(similarity, 4)
                    })

# Convert results to DataFrame for better viewing
similarity_df = pd.DataFrame(similarity_results)

# Display similarity DataFrame
print(similarity_df)

# Save the results to an Excel file
output_path = "C:/Users/Nastaran/Desktop/UniversityConstructor/semester4/Final 08042025/Data/sentence_similarity_results.xlsx"
similarity_df.to_excel(output_path, index=False)
```

```python
import os
import docx
import re
import pandas as pd
from gensim.models import Word2Vec
from sklearn.metrics.pairwise import cosine_similarity

# Function to read content from a docx file
def read_docx(file_path):
    doc = docx.Document(file_path)
    full_text = []
    for para in doc.paragraphs:
        full_text.append(para.text)
    return '\n'.join(full_text)

# Function to preprocess text (remove punctuation and convert to lowercase)
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text

# Read all DOCX files and preprocess text
directory = 'C:/Users/Nastaran/Desktop/UniversityConstructor/semester4/Final 08042025/Data/Similarity'
texts = []
filenames = []

for filename in os.listdir(directory):
    if filename.endswith(".docx"):
        file_path = os.path.join(directory, filename)
        text = read_docx(file_path)
        texts.append(preprocess_text(text))  # Preprocess the text
        filenames.append(filename)

# Tokenize the texts into words
tokenized_texts = [text.split() for text in texts]

# Train a Word2Vec model on the tokenized texts
model = Word2Vec(tokenized_texts, vector_size=100, window=5, min_count=1, workers=4)

# Compute cosine similarity between all document pairs
similarity_results = []

# Calculate similarity between documents
for i, filename1 in enumerate(filenames):
    for j in range(i + 1, len(filenames)):
        text1 = texts[i]
        text2 = texts[j]

        # Tokenize the texts
        tokens1 = text1.split()
        tokens2 = text2.split()

        # Calculate word2vec vectors for each document by averaging the word vectors of the words in the document
        vectors1 = [model.wv[word] for word in tokens1 if word in model.wv]
        vectors2 = [model.wv[word] for word in tokens2 if word in model.wv]

        if vectors1 and vectors2:
            avg_vector1 = sum(vectors1) / len(vectors1)
            avg_vector2 = sum(vectors2) / len(vectors2)

            # Compute cosine similarity between the average vectors
            similarity = cosine_similarity([avg_vector1], [avg_vector2])[0][0]

            # If similarity is above a threshold, store the result
            if similarity > 0.5:  # You can adjust the threshold
                similarity_results.append({
                    'File 1': filenames[i],
                    'File 2': filenames[j],
                    'Similarity': round(similarity, 4)
                })

# Convert the results to a DataFrame
similarity_df = pd.DataFrame(similarity_results)

# Display the similarity DataFrame
print(similarity_df)

# Save the results to an Excel file
output_path = "C:/Users/Nastaran/Desktop/UniversityConstructor/semester4/Final 08042025/Data/word2vec_similarity_results.xlsx"
similarity_df.to_excel(output_path, index=False)
```
